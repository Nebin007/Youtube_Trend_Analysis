```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]

numcolist = ['category_id','views','likes','dislikes']
CAdat = pd.read_csv("Cleaned/CAvideos.csv")
GBdat = pd.read_csv("Cleaned/GBvideos.csv")
USdat = pd.read_csv("Cleaned/lSvideos.csv")
categories = pd.read_csv("categories.csv")

for co in numcolist:
    CAdat[co]= pd.to_numeric(CAdat[co], errors='coerce')
    CAdat[co] = CAdat[co].fillna(0)
    GBdat[co]= pd.to_numeric(GBdat[co], errors='coerce')
    GBdat[co] = GBdat[co].fillna(0)
    USdat[co]= pd.to_numeric(USdat[co], errors='coerce')
    USdat[co] = USdat[co].fillna(0)
CAdat['title'] = CAdat['title'].fillna('-')
GBdat['title'] = GBdat['title'].fillna('-')
USdat['title'] = USdat['title'].fillna('-')
CAdat['tags'] = CAdat['tags'].astype(str)
GBdat['tags'] = GBdat['tags'].astype(str)
USdat['tags'] = USdat['tags'].astype(str)
CAdat['tags'] = CAdat['tags'].fillna('-')
GBdat['tags'] = GBdat['tags'].fillna('-')
USdat['tags'] = USdat['tags'].fillna('-')
CAdat['description'] = CAdat['description'].fillna('-')
GBdat['description'] = GBdat['description'].fillna('-')
USdat['description'] = USdat['description'].fillna('-')
```

```python
import collections
tagsgen = []
CAtaglist = CAdat.loc[CAdat['category_id'] == 806].values.tolist()
convertlist = [str(i) for i in CAtaglist]
tagstr = ",".join(convertlist)
tagstr = tagstr.replace("'","")
tagstr = tagstr.replace("[","")
tagstr = tagstr.replace("]","")
tagstr = tagstr.replace("\\","")
tagsgen.extend(tagstr.split(","))
GBtaglist = GBdat.loc[GBdat['category_id'] == 806].values.tolist()
convertlist = [str(i) for i in GBtaglist]
tagstr = ",".join(convertlist)
tagstr = tagstr.replace("'","")
tagstr = tagstr.replace("[","")
tagstr = tagstr.replace("]","")
tagstr = tagstr.replace("\\","")
tagsgen.extend(tagstr.split(","))
UStaglist = USdat.loc[USdat['category_id'] == 806].values.tolist()
convertlist = [str(i) for i in UStaglist]
tagstr = ",".join(convertlist)
tagstr = tagstr.replace("'","")
tagstr = tagstr.replace("[","")
tagstr = tagstr.replace("]","")
tagstr = tagstr.replace("\\","")
tagsgen.extend(tagstr.split(","))
frequency = dict(collections.Counter(tagsgen))
top15 = dict(collections.Counter(frequency).most_common(18))
del top15[' False']
del top15[' "']
del top15[' 806.0']
plt.rcParams["figure.figsize"] = (15,5)
X = top15.keys()
Y = top15.values()
X_axis = np.arange(len(X))
  
plt.bar(X_axis, Y, 0.4, label = 'Frequency')
plt.xticks(X_axis, X, rotation=45)
plt.xlabel("Tagwords")
plt.ylabel("Frequency")
plt.legend()
```

<p> Now Let's check the keywords of each video, for that I will be implementing keyword extraxction using python.<br>
For that I will be using title and description of all trending videos.</p>
<p>To achieve that easily I have put all the title and description into a text file. I will be loading all of the zones to find <br>
the best suitable ones.</p>


```python
import yake
alltitle = open("Cleaned/title/catitle.txt", "r").read() +","+ open("Cleaned/title/ustitle.txt", "r").read()+","+open("Cleaned/title/gbtitle.txt", "r").read()
kw_ext = yake.KeywordExtractor(top = 15, stopwords=None)
titlekeywords = kw_ext.extract_keywords(alltitle)
print(titlekeywords)
```

<p>For finding the key words as above, I made use of yake.</p>
<p>Yake is a lightweight, unsupervised automatic keyword extraction method that relies on <br>statistical text features extracted from individual documents to identify the most relevant keywords in the text. This system does not need to be trained on a particular set of<br> documents, nor does it depend on dictionaries, text size, domain, or language. Yake defines a set of five features capturing keyword characteristics which are heuristically <br>combined to assign a single score to every keyword. The lower the score, the more significant the keyword will be.</p>

<p> But i would prefer text rank, of summa rather than yake. Since It shows the exact keyword. I used yake, because <br>
text-rank is not suitable for a large texts </p>

<p>To analize this lets plot a graph.</p>

```python
plt.rcParams["figure.figsize"] = (15,5)
X = []
Y = []
for kw, score in titlekeywords:
    X.append(kw)
    Y.append(score)
X_axis = np.arange(len(X))
  
plt.bar(X_axis, Y, 0.4, label = 'Keyword scores')
plt.xticks(X_axis, X, rotation=45)
plt.xlabel("Keywords")
plt.ylabel("Scores")
plt.legend()
```

<p>As mentioned above, the lesser the score, the more significant the values.</p>
